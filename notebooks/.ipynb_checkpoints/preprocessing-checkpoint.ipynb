# Group levels of `Standing`

To start, we'll examine the `Standing` column for the 2019-2020 budget year.

```python
rso_2019 = rso_data[rso_data['Year'] == 2019]
```

```python
sns.set(rc={'figure.figsize':(20,8)})

fig, ax = plt.subplots(1,2)

sns.boxplot(x='Standing', data=rso_2019, ax=ax[0]).set_title('Boxplot of Standing for 2019-2020')
sns.distplot(rso_2019['Standing'], bins=np.arange(135), ax=ax[1]).set_title('Histogram of Standing for 2019-2020')

fig.savefig('C:\\Users\\scher\\school\\senior-thesis\\images\\bp_hist_standing_2019.png')
```

The boxplot and histogram both show that `Standing` is extremely skewed to the right -- that most clubs tend to be only a couple years old. We'll see if this trend exists in other years. 

```python
sns.set(rc={'figure.figsize':(20,20)})

# The years we want to plot can be edited by adding to the following array. That said, the figure size (in the line above)
# must be adjusted accordingly.
years = [2017, 2018, 2020]

fig, ax = plt.subplots(len(years),2)

for i, year in enumerate(years):
    data = rso_data[rso_data['Year']==year]
    
    sns.boxplot(x='Standing', data=data, ax=ax[i][0]).set_title(f'Boxplot of Standing for {year}-{year+1}')
    sns.distplot(data['Standing'], bins=np.arange(135), ax=ax[i][1]).set_title(f'Histogram of Standing for {year}-{year+1}')
    
fig.savefig('C:\\Users\\scher\\school\\senior-thesis\\images\\bp_hist_standing_2017_18_20.png')
```

Although `Standing` certainly reflects an overall gradual increase over time, RSO standing, in general, is still extremely skewed to the left. This could be attributed to the fact that new clubs are be volatile -- they may only last a few years before being disbanded. 

Bins for `Standing` must be created with this in mind. 

```python
# Create bins for every 10th percentile based on budget year, which accounts for specific ASUC guidelines for funding clubs
# with a standing of 3 or lower.
bin_vals_dict = {}

for y in rso_data['Year'].unique():
    data_by_year = rso_data[rso_data['Year']==y]
    bins = np.unique([np.round(np.percentile(data_by_year['Standing'], p)) for p in np.arange(0,100,10)])
    
    bin_vals_dict[y] = bins
    
bin_vals_dict
```

```python
def add_standing_bins(row, rso_data, bin_vals_dict):
    year = row['Year']
    standing = row['Standing']
    
    bins_by_year = bin_vals_dict[year]
    bin_num = np.argmin(abs(bins_by_year-standing))
    
    return bin_num
```

```python
rso_data['Binned Standing'] = rso_data.apply(lambda r: add_standing_bins(r, rso_data, bin_vals_dict), axis=1)
rso_data
```

# Normalization

```python
norm_rso_data = rso_data.copy(deep=True)
```

## Normalizing values by year

```python
def y_norm(data, row):
    year = row[['Year']].values[0]
    alloc = row[['Allocation']].values[0]
    
    year_mean = np.mean(data[data['Year'] == year]['Allocation'])
    year_std = np.std(data[data['Year'] == year]['Allocation'])
                      
    return (alloc - year_mean) / year_std
```

```python
norm_rso_data['norm_by_y'] = norm_rso_data.apply(lambda r: y_norm(rso_data, r), axis=1)
norm_rso_data
```

## Normalizing values by year and designation

```python
def yd_norm(data, row):
    year = row[['Year']].values[0]
    desig = row[['Designation']].values[0]
    alloc = row[['Allocation']].values[0]
    
    year_mean = np.mean(data[(data['Year'] == year) 
                             & (data['Designation'] == desig)]
                        ['Allocation'])
    year_std = np.std(data[data['Year'] == year]['Allocation'])
                      
    return (alloc - year_mean) / year_std
```

```python
norm_rso_data['norm_by_yd'] = norm_rso_data.apply(lambda r: yd_norm(rso_data, r), axis=1)
norm_rso_data
```

## Normalize values by year and binned standing

```python
def ys_norm(data, row):
    year = row[['Year']].values[0]
    standing = row[['Binned Standing']].values[0]
    alloc = row[['Allocation']].values[0]
    
    year_mean = np.mean(data[(data['Year'] == year) 
                             & (data['Binned Standing'] == standing)]
                        ['Allocation'])
    year_std = np.std(data[data['Year'] == year]['Allocation'])
                      
    return (alloc - year_mean) / year_std
```

```python
norm_rso_data['norm_by_ys'] = norm_rso_data.apply(lambda r: ys_norm(rso_data, r), axis=1)
norm_rso_data
```

## Normalize values by year, designation, and binned standing

```python
def yds_norm(data, row):
    year = row[['Year']].values[0]
    desig = row[['Designation']].values[0]
    standing = row[['Binned Standing']].values[0]
    alloc = row[['Allocation']].values[0]
    
    year_mean = np.mean(data[(data['Year'] == year) 
                             & (data['Designation'] == desig) 
                             & (data['Binned Standing'] == standing)]
                        ['Allocation'])
    year_std = np.std(data[data['Year'] == year]['Allocation'])
                      
    return (alloc - year_mean) / year_std
```

```python
norm_rso_data['norm_by_yds'] = norm_rso_data.apply(lambda r: yds_norm(rso_data, r), axis=1)
norm_rso_data
```
